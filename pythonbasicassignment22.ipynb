```python
#4)
X = "ineuron"
def func1():
    global X
    X = "NI!"
    print(X)
```

```python
func1()
print(X)
```

```python
#Global keyword allows a variable to be accessible in the current scope.when we are using global keyword inside the function
#func1 directly access the variable in X in global scope and changes its value to NI!.so, the output of the code is NI!
```

```python
#6)
def func():
    X = 'NI!'
    def nested():
        nonlocal X
        X = 'Spam'
    nested()
    print(X)
func()
```

```python
# nonlocal keyword in python is used to declare a variable as not local.
#Statement X = "Spam" is modified in the globalscope.So,output of print(X) statement is Spam
```
