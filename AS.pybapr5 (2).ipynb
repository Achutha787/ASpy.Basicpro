```python
#2)python program to find hcf
x = int(input("enter the first number\n"))
y = int(input("enter the second number\n"))
if x > y:
  x, y = y, x
for i in range(1,x+1):
  if x%i == 0 and y%i == 0:
    hcf = i
print("HCF of", x, "and", y, "is:", hcf)
```

```python
#3)python program to convert decimal to binary,octal and hexadecimal
def decimal_into_binary(decimal_1):  
    decimal = int(decimal_1)   
    print ("The given decimal number", decimal, "in Binary number is: ", bin(decimal))  
    
def decimal_into_octal(decimal_1):  
    decimal = int(decimal_1)  
    print ("The given decimal number", decimal, "in Octal number is: ", oct(decimal))  
  
def decimal_into_hexadecimal(decimal_1):  
    decimal = int(decimal_1)    
    print ("The given decimal number", decimal, " in Hexadecimal number is: ", hex(decimal))  
    
# Driver program  
decimal_1 = int (input (" Enter the Decimal Number: "))  
decimal_into_binary(decimal_1)  
decimal_into_octal(decimal_1)  
decimal_into_hexadecimal(decimal_1)  
```

```python
#5)Python program to make a simple calculator with 4 basic mathematical operations
def addition(x,y):
    return("sum",x+y)
def substraction(x,y):
    return("substraction",x-y)
def multiply(x,y):
    return("multiply",x*y)
def division(x,y):
    return("division",x/y)
print("Please select operation -\n" \
        "1. addition\n" \
        "2. Subtraction\n" \
        "3. Multiply\n" \
        "4. Division\n")
select= int(input("select operations from 1,2,3,4"))
x = int(input( "enter first number\n"))
y = int(input("enter second number\n"))
if select == 1:
    print(x,"+",y,"=",addition(x,y))
elif select == 2:
    print(x,"-",y,"=",substraction(x,y))
elif select == 3:
    print(x,"*",y,"=",multiply(x,y))
elif select == 4:
    print(x,"/",y,"=",division(x,y))
else:
    print("invalid input ")
    
  
```
