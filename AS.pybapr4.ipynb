```python
#1)python program to print factorial of a number
num = int(input("enter the number"))
fact = 1
for i in range(2,num+1):
    fact = fact * i
print("factorial of no. ",num,"is",fact)
```

```python
#3)python program to print Fibonacci series
nterms = int(input("enter no. of fibonacci sequence needed"))
count = 0
n1,n2 = 0,1
if nterms <= 0:
    print("pls enter a positive integer")
elif nterms == 1:
    print("fibonacci sequence upto",nterms,":")
    print(n1)
else:
    print("fibonacci sequence :")
    while count < nterms:
        print(n1)
        nth = n1 +n2
        n1 = n2
        n2 = nth
        count= count+1
    
```

```python
#4)Python program to check for an armstrong number
A = int(input("enter your no."))
order = len(str(A))
sum = 0
temp = A
while temp> 0:
    digit =temp % 10
    sum+= digit** order
    temp//= 10
if A == sum:
    print(A, " is armstrong number")
else:
    print(A, " is not armstrong number")
```

```python
#5)python program to print Armstrong number in interval
A = int(input("enter the lower number interval"))
B = int(input("enter the upper number interval"))
for i in range(A,B+1):
    order = len(str(i))
    sum = 0
    temp = i
    while temp > 0:
        digit = temp% 10
        sum+= digit** order
        temp//= 10
    if i == sum:
        print(i)
```
