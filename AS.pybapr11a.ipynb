```python
#1)python program to find words greater than length k
def string_k(k, str):
    string = []
    text = str.split(" ")
    for x in text:
        if len(x)> k:
            string.append(x)
    return string
k = 3
string="i am a data scientist"
print(string_k(k,string))
```

```python
#5)python program to find uncommon words from 2 strings
firststring="my brother is living in hyderabad"
secondstring = "my sister is living in banglore"
A = firststring.split()
B = secondstring.split()
UC = " "
for i in A:
    if i not in B:
        UC = UC +" "+i
for j in B:
    if j not in A:
        UC = UC +" "+j
print(UC)
```
